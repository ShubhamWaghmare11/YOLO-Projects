```python
from ultralytics import YOLO
```

```python
import matplotlib.pyplot as plt
model = YOLO('../Yolo-Weights/yolov8n.pt')
```

```python
results = model("images/img1.jpg")


img = results[0].plot()
```

```python
import cv2

plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.axis('off');
```
